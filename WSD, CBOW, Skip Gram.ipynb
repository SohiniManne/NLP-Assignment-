```python
# Install gensim
!pip install gensim

# Import libraries
from gensim.models import Word2Vec

# Anirudh songs and movies dataset
sentences = [
    ["anirudh", "composed", "bgm", "for", "vikram"],
    ["kaavaalaa", "song", "is", "trending", "worldwide"],
    ["anirudh", "and", "rajinikanth", "collaborated", "in", "jailer"],
    ["beast", "arabic", "kuthu", "became", "viral", "hit"],
    ["anirudh", "is", "the", "top", "music", "composer", "in", "kollywood"],
    ["thalapathy", "vijay", "and", "anirudh", "make", "blockbuster", "movies"],
    ["varisu", "master", "leo", "are", "anirudh", "movies"],
    ["vaathi", "coming", "is", "a", "mass", "song", "from", "master"],
    ["kannaana", "kanney", "is", "emotional", "song", "from", "viswasam"],
    ["anirudh", "background", "score", "elevates", "every", "scene"],
    ["naan", "ready", "is", "peppy", "song", "by", "anirudh"],
    ["anirudh", "ravichander", "is", "young", "talented", "composer"],
    ["leo", "movie", "songs", "composed", "by", "anirudh", "are", "superhit"],
    ["jailer", "bgm", "by", "anirudh", "is", "thrilling", "and", "powerful"],
]

# CBOW Model (sg=0 means CBOW)
model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    sg=0,
    epochs=300
)

# Words similar to anirudh
print("Words similar to 'anirudh':")
print(model.wv.most_similar("anirudh"))

print("\nWords similar to 'song':")
print(model.wv.most_similar("song"))

print("\nWords similar to 'bgm':")
print(model.wv.most_similar("bgm"))

# Word vector
print("\nVector for 'anirudh':")
print(model.wv["anirudh"])

# Similarity between words
print("\nSimilarity between 'anirudh' and 'composer':")
print(model.wv.similarity("anirudh", "composer"))

print("\nSimilarity between 'vikram' and 'jailer':")
print(model.wv.similarity("vikram", "jailer"))

# How CBOW works - plain text explanation
print("\n--- How CBOW Works ---")
print("Context: [anirudh, composed, for, vikram] -> Predicts: BGM")
print("CBOW takes surrounding words and predicts the center word!")
```

```python
import nltk
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

# Example 1 - BANK
sentence1 = "I went to the bank to deposit money"
sentence2 = "I sat by the bank of the river"
word = "bank"

sense1 = lesk(word_tokenize(sentence1), word)
sense2 = lesk(word_tokenize(sentence2), word)

print("=== BANK EXAMPLE ===")
print(f"Sentence 1: {sentence1}")
print(f"Sense found: {sense1}")
print(f"Definition: {sense1.definition()}")

print(f"\nSentence 2: {sentence2}")
print(f"Sense found: {sense2}")
print(f"Definition: {sense2.definition()}")

# Example 2 - ROCK (Anirudh context)
sentence3 = "anirudh plays rock music in his concerts"
sentence4 = "the rock fell down from the mountain"
word2 = "rock"

sense3 = lesk(word_tokenize(sentence3), word2)
sense4 = lesk(word_tokenize(sentence4), word2)

print("\n=== ROCK EXAMPLE ===")
print(f"Sentence 3: {sentence3}")
print(f"Sense found: {sense3}")
print(f"Definition: {sense3.definition()}")

print(f"\nSentence 4: {sentence4}")
print(f"Sense found: {sense4}")
print(f"Definition: {sense4.definition()}")
```

```python
!pip install gensim
from gensim.models import Word2Vec

# Anirudh songs and movies dataset
sentences = [
    ["anirudh", "composed", "bgm", "for", "vikram"],
    ["kaavaalaa", "song", "is", "trending", "worldwide"],
    ["anirudh", "and", "rajinikanth", "collaborated", "in", "jailer"],
    ["beast", "arabic", "kuthu", "became", "viral", "hit"],
    ["anirudh", "is", "the", "top", "music", "composer", "in", "kollywood"],
    ["thalapathy", "vijay", "and", "anirudh", "make", "blockbuster", "movies"],
    ["varisu", "master", "leo", "are", "anirudh", "movies"],
    ["vaathi", "coming", "is", "a", "mass", "song", "from", "master"],
    ["kannaana", "kanney", "is", "emotional", "song", "from", "viswasam"],
    ["anirudh", "background", "score", "elevates", "every", "scene"],
    ["naan", "ready", "is", "peppy", "song", "by", "anirudh"],
    ["anirudh", "ravichander", "is", "young", "talented", "composer"],
    ["leo", "movie", "songs", "composed", "by", "anirudh", "are", "superhit"],
    ["jailer", "bgm", "by", "anirudh", "is", "thrilling", "and", "powerful"],
]

# Skip-gram Model (sg=1)
model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    sg=1,        # 1 = Skip-gram
    epochs=300
)

# Words similar to anirudh
print("Words similar to 'anirudh':")
print(model.wv.most_similar("anirudh"))

print("\nWords similar to 'song':")
print(model.wv.most_similar("song"))

print("\nWords similar to 'bgm':")
print(model.wv.most_similar("bgm"))

# Word vector
print("\nVector for 'anirudh':")
print(model.wv["anirudh"])

# Similarity between words
print("\nSimilarity between 'anirudh' and 'composer':")
print(model.wv.similarity("anirudh", "composer"))

print("\nSimilarity between 'vikram' and 'jailer':")
print(model.wv.similarity("vikram", "jailer"))

# How Skip-gram works
print("\n--- How Skip-gram Works ---")
print("Center word: BGM")
print("Predicts context: [anirudh, composed, for, vikram]")
print("Skip-gram takes center word and predicts surrounding words!")
```
